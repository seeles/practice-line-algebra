```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 1, -1]
y = [2, 4, 2, 1]

fig = plt.figure()
plt.axhline(y = 0, c='black')
plt.axvline(x = 0, c='red')

plt.plot(x[:2],y[:2], x[2:], y[2:], c='blue')
plt.draw()
```

```python
x = [-2, 2, -2, 2]
y = [-4, 4, 0.5, 2.5]

fig = plt.figure()
plt.axhline(y=0, c='black')
plt.axvline(x=0, c='black')

plt.plot(x[:2], y[:2], x[2:], y[2:])

plt.draw()
```
